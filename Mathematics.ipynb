# Exploration of the mathematics of the Super probability Plots, Beta-Binomial and Dirichlet-Multinomial Distributions

In this chapter, we explore the mathematic involved in beta binomial and Dirichlet multinomial distributions and relate them by using the concept of the Super Probability Minima plot developed in the previous chapter. This concept is important as a generalisation to model the lower limit of prevalence levels for gene expression as they not only vary for one particular allelic variant but also considering multiple categories of alleles for genes. Here, we aim to establish the deductive structure for inferring population prevalence for different levels of disease expression. If it is possible to discover that the model can simulate different ranges of prevalence, this would provide external validity for the gene expresion model. So far, we are developing gene expression of multiple genes that are independent of each other. Following this step, we will generate a model of a Dirichlet Multinomial model, that will provide the basis for a Hierarchical Dirichlet model that will take into account dependent interactions such as feed forward and feedback mechanisms.
We commence the mathematical foundations with the Beta function and ts derivation.

## Beta function as a Probability Distribution

With the beta-binomial distribution we seek to create a Bayesian prior that is updated as a posterior depending on new data. It is assumed that gene expression is constantly being updated due to regulatory and gene-gene interactions. At present, we are considering only one level of such interaction as a level or 1D array of a Binomial Distribution. Instead of having a set prior probability of , say dominant expression, the beta-binomial distribution permits exploration of a range of such probabilities. It is a continuous distribution compared to the binomial distirbution which is a iscrete distribution of probability.
  
We start with considering a probability distribution where x is probability as a variable, whereas the exponents a and b are constants.

### Beta Function- Derivation

Here we commence by defining a continuous probability function where x is a variable probability and exponents a and b are equivalent to k and n-k. The probability can vary from 0 to 1 where p + (1-P) = 1. For any prior probability p, the product of p and its conjugate (1-p) would be summed over a binomial distribution of probabilities which add up to 1. Sum of probabilities could be cosidered an integral in the limit for large n.

Take,
  
$$ 
  \int_0^1 x^{a}(1-x)^{b}\,dx   
$$

Let $u=(1-x)^{b}  \Rightarrow   du=b(1-x)^{b-1}(-1)dx$
  
and $dv=x^{a}dx  \Rightarrow   v= \frac{x^{a+1}}{a+1}$

**(for example, considering a group of genes n, a = k and b = (n-k))**

**Integration by Parts**

We apply the identities above to define the rule for integration by parts:


$$ 
  \int_a^ b u\,dv = |uv(b)-uv(a)| - \int_a^ b v\,du
$$  



**Integrate by parts repeatedly to remove $(1-x)^b$ term.**

The following recursive process of repeated integration by parts results in elimination of the $(1-x)^b$ term and results in a factorial solution.

$$ 
  \int_0^1 x^{a}(1-x)^{b}\,dx
$$

= $(1-x)^{b}\frac{x^{a+1}}{a+1}|_0^1 - \int_0^1 b(1-x)^{b-1}(-1)\frac{x^{a+1}}{a+1}\,dx$

= $\frac{b}{a+1}\int_0^1 x^{a+1}(1-x)^{b-1}\,dx$

= $\frac{b(b-1)}{(a+1)(a+2)}\int_0^1 x^{a+2}(1-x)^{b-2}\,dx$

...

= $\frac{b!}{(a+1)(a+2)...(a+b)}\int_0^1 x^{a+b}\,dx$

= $\frac{b!a!}{(a+b)!}\int_0^1 x^{a+b}\,dx$

= $\frac{b!a!}{(a+b)!}\frac{x^{a+b+1}}{a+b+1}|_0^1$

= $\frac{b!a!}{(a+b)!}\frac{1}{a+b+1}$

= $\frac{b!a!}{(a+b+1)!}$

**Beta Function, B(𝛼,𝛽) can be expressed in terms of th gamma (𝛤) function or factorials**
  
Let 𝛼 = a+1 and 𝛽 = b+1,

$\Rightarrow$ 𝛤(𝛼) = 𝛤(a+1) = a! and 𝛤(𝛽) = 𝛤(b+1) = b!

$\Rightarrow$ (a+b+1)! = 𝛤(a+b+2) = 𝛤(𝛼+𝛽)

$\Rightarrow \int_0^1 x^{a}(1-x)^{b}\,dx = \int_0^1 x^{𝛼-1}(1-x)^{𝛽-1}\,dx = \frac{𝛤(𝛼)𝛤(𝛽)}{(𝛤(𝛼+𝛽)}$ 
  
=$\frac{a!b!}{(a+b+1)!}$


## Application of Bayesian inference to derive the Beta-Binomial Distribution

**Bayes Rule for continuous variable P and discrete Y (gene numbers)**

The beta distribution to be described below is considered a conjugate distribution for the binomial distribution as it permits choice of different value of probability for succes (dominant expresion) in a continuous range from 0 to 1. It permits Bayesian inference and continuous updating of probabilities of expression as a the data changes.
  
**$ 𝝧_{AB}(P|Y) = f(Y|P)⋅f_{A}(P)/P(Y)$**

𝜃 = Updated Belief (estimate) - a beta-distribution
  
f(Y|P) = Probability distribution of gene configurations - a binomial distribution
  
$f_{A}(P)$ = Prior belief about probability (e.g. of dominant expression) - a beta distribution
  
P(Y) = Probability of data after collection (i.e. frequency estimates)

**Beta Function**

$\int_0^1 x^{𝛼-1}(1-x)^{𝛽-1}\,dx = B(𝛼,𝛽)$
  
$\Rightarrow$ $\int_0^1(\frac{1}{B(𝛼,𝛽)}{x^{𝛼-1}(1-x)^{𝛽-1}}\,dx$ = 1

f(x) = $\int_0^1(\frac{1}{B(𝛼,𝛽)}{x^{𝛼-1}(1-x)^{𝛽-1}}\,dx$ 0<x<1, is the **Beta Distribution**, 

where B(𝛼,𝛽) is a constant

**Beta-Binomial Distribution**

this is the posterior (Bayesian) esitmate of probabilities for a binomial distribution where, 

$ 𝝧_{AB}(P|Y)  = f(Y|P)⋅f_{A}(P)/P(Y)$    
               = ($^{n}C_{k}$) $p^{k}(1-p)^{n-k}$ ${cp^{𝛼-1}(1-p)^{𝛽-1}}/{P(Y)}$
                                                                    (here P(Y) is a constant independent of P)        
               = Constant⋅($p^{k+𝛼-1}(1-p)^{n-k+𝛽-1}$)   (c and P(Y) are constants)    
               = Beta(p, k+𝛼, n-k+𝛽)

  
As the distribution could be considered a sum of probabilities as seen in the binomial distribution of gene configuration probabilities, the compound distribution can be considered as an integral,

$\int_0^1(\frac{1}{B(𝛼,𝛽)}{p^{k}}{(1-p)^{n-k}}{x^{𝛼-1}(1-x)^{𝛽-1}}\,dx$ 0<x<1

## Beta Binomial Distribution and Super Probability Minima Plot - Mathematics

Following is the general beta-binomial equation for calculating the probabilities of different gene configurations.

$^{n}C_{k}$ [Γ(k+α) Γ(n -k +β)/ Γ(n+ α +β)]/ [Γ(α) Γ(β)/ Γ(α+β)]
where,

Γ(x)  = x-1!

For k=n/2, the equation is,

$^{n}C_{n/2}$ [Γ(n/2+α) Γ(n/2 +β)/ Γ(n+ α +β)]/ [Γ(α) Γ(β)/ Γ(α+β)]

The above code uses the equation above to calculate the minimum of the maximum probability curve (the super-probability curve) for any n and then plots those points for a range of n (above it is plotted to n = 50).

Later, the mathematical insights that this approximation brings to interpretation of the Dirichlet-Multinomial Model will be discussed.


## Generalisation of the beta-binomial distribution to the Dirichlet-multinomial distribution

The next step was to generalise the idea of the balanced expression of gene-configurations under the beta-binomial distribution to the Dirichlet-multinomial distribution.
The code computes the probability of observing a perfectly balanced count vector under a symmetric Dirichlet-multinomial distribution (where all concentration parameters are set to 1). For a sample N, the individual data points would be spread among and assigned to k categories. The key equation is:
Probability of a Specific Count Vector
For a total count (n) and (k) categories, the probability of any specific count vector ((x_1, x_2, …, x_k)) (where (x_i >/= 0) and 𝛴ki=1 xi = n)) is:

P(x) = [𝛤(n + 1)⋅𝛤(k)]/𝛤(n + k)
 
where 𝛤is the gamma function. For integer inputs:
•	𝛤(n + 1) = n!) (factorial of (n)),
•	𝛤(k) = (k-1)!) (since (k) is an integer),
•	𝛤(n + k) = (n + k - 1)!).
Thus, the equation simplifies to:

P(x) = [n!⋅(k-1)!]/(n+k-1)!

**Special Case: "Balanced" Vector**
The code implicitly assumes the balanced vector ((x_1 = x_2 = … = xk = n/k)) exists (i.e., (n/k) is an integer). In this case:
P_balanced = [n! × (k-1)!]/[(n+k-1)!]

If (n/k) is not an integer, the probability of a perfectly balanced vector is (0), but the code does not check this and computes the formula regardless.

n/k replaces k=n/2 in the beta-binomial case where k is the number of dominantly expressed genes in a combination n. The two k’s should not be mixed up. n/k here represents the number of categories not the number of expressed genes of a particular category. Instead k could be replaced by c (for category), thus n/c = number of expressed genes of a particular category (e.g. dominant, recessive or intermediate, for balanced expression it could be expected that n/c = 0.33 but the assumption is for  integer values – so more like ,if there are 12 genes 4 belong to one category). This is in contrast to the concentration parameter 𝛼 that determines how closely the data is concentrated for a particular category. The model uses this category parameter to represent levels of gene expression (more accurately categories of gene expression – multi-allelic vs bi-allelic). When there are two levels (highly expressed and minimally expressed) c=2 and the distributions resemble beta-binomial whereas, for 3 levels (high, intermediate, low) c=3 and the curve asymptote is lower. Similarly with c = 4 and 5.


## Why is the Dirichlet multinomial distribution a generalisation of the beta-binomial distribution in the case of balanced probability construct?

The probability equation above generalizes the beta-binomial super probability minima plot to (k > 2) categories. Here's why:
Beta-Binomial Superprobability Minima (for (k=2))
•	In the beta-binomial model (2 categories), the symmetric case (concentration parameters (𝛼 = 𝛽 = 1)) gives a uniform prior. (In fact, in the code that explores this 𝛼 = 𝛽 = n/2 (assuming that this is an integer) and this is discussed below).
•	If we consider beta-binomial probability equation as  
 $^{n}C_{k}$[Γ(k+α) Γ(n -k +β)/ Γ(n+ α +β)]/ [Γ(α) Γ(β)/ Γ(α+β)]  
when k=n/2, and (𝛼 = 𝛽 = 1)) ,   
 $P_{balanced} = ^{n}C_{n/2}[Γ((n/2)+1)⋅Γ((n-n/2)+1)/Γ(n+ 2)]/[Γ(1)⋅Γ(1)/Γ(1+1)]$   **([Γ(1)⋅Γ(1)/Γ(1+1)]=1])**  
= $^{n}C_{n/2}[Γ((n/2)+1)⋅Γ(n-n/2+1)/Γ(n+ 2)] = [n!/(n/2)!⋅(n/2)!][((n/2)!⋅(n/2)!)/n+1!]$  
= n!/n+1! = 1/n+1    
•	Taking the generalized equation for the perfectly balanced Dirichlet multinomial vector,The minimum probability over all possible counts x ∊ {0, 1, … , n}) is:

 
min(x) P(X = x) = [n!⋅(k-1)!]/(n+k-1)! = [n!⋅1!]/(n+1)! = 1/n+1

(matches the formula for k=2 above).
•	This is the "superprobability minima" curve for (k=2).

•	If 𝛼 = 𝛽 = n/2 rather than 1, then the value of minP(X=x) can be significantly different. This is what we find in the code that calculates the minima with n=k/2 (i.e. for n = 10 genes, 𝛼 = 𝛽 = n/2+1 = 6; then the equation for Pbalanced is,

$P_{balanced} = ^{10}C_{5}$[Γ6+5)⋅Γ(6+5)]/Γ(22)]/[Γ(6)⋅Γ(6)/Γ(12)]   
= 10!/5!5! [10!10!/21!]/5!5!/11!)     
= 10!/5!5! [10!10!/21!][11!/5!5!]  
= 10!10!10!11!/5!5!5!5!21!  
= (3⋅10!)⋅11! /(4⋅5!)⋅21!  
≈ 1.91 x $10^{27}/ 1.059 x 10^{28}$  
= 0.180 (whereas 1/n+1 = 1/11 = 0.091)

Using the beta-binomial probability distribution directly, adjusting it according to empirical states (e.g. choosing the probability of gene expression for 10 genes for their various combinations, requires considering different values of 𝛼 and 𝛽 according to proportion of gens of each category in each combination of n genes. For 10 genes, the number of configurations is 11, and if 𝛼 is 9 and 𝛽, 3, the expected dominant gene expression probability is 0.75 (Mendelian). As seen above, changing the hyperparameters according to expected frequency in gene configuration can alter the actual probability of this combination. However, the Stirling approximation brings the equation close to this empirically expected pattern. In the figure below, the difference between the empirical curve, the deduced curve (assuming 𝛼=𝛽=1) and the Stirling approximation are displayed. The optimized beta-binomial is empirically adjusted while the blue curve is the one deduced with 𝛼 =𝛽 =1. The Stirling asymptotic curve Probability/prevalence = √1/𝜋n demonstrates that it approximates (at n ≫ 5) what would be expected of empirical data distribution. The derivation of √1/𝜋n from an approximation of the alpha-multinomial is described below.

Below, we provide the code that generates the figure above. In addition, the code also plots a form of Stirling Approximation for the equivalent Dirichlet curves. The deductive justification is covered following the code.  

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gammaln, binom, beta
from scipy.stats import betabinom

# ==============================
# 1. CORRECTED Beta-Binomial Optimized Curve
# ==============================
def optimized_beta_binomial(n):
    """Compute maximum achievable probability at k=n/2 with optimized prior"""
    k = n // 2
    a = b = k + 1  # Optimized prior parameters
    
    # Direct calculation using beta-binomial PMF
    return betabinom.pmf(k, n, a, b)

# ==============================
# 2. Dirichlet-Multinomial Functions
# ==============================
def exact_dirichlet_multinomial(n, k):
    """Exact probability for balanced case with symmetric prior (α_i=1)"""
    if n % k != 0:  # Only defined when n/k is integer
        return 0
    return np.exp(gammaln(n+1) + gammaln(k) - gammaln(n+k))

def stirling_dirichlet_multinomial(n, k):
    """Stirling approximation for large n"""
    return np.sqrt(k) * (2 * np.pi * n)**((1 - k)/2)

# ==============================
# 3. Parameter Setup
# ==============================
# Gene count ranges
n_beta = np.arange(5, 51, 2)  # For beta-binomial (5-50 genes)
n_dirichlet = np.arange(5, 201)  # For Dirichlet-multinomial (5-200 genes)

# Categories
categories = [2, 3, 4, 5]

# Prevalence thresholds
thresholds = [0.5, 0.2, 0.1, 0.05, 0.01]
threshold_colors = ['green', 'blue', 'orange', 'red', 'purple']
threshold_labels = [
    'Common phenotypes (>50%)',
    'High prevalence disorders (20%)',
    'High prevalence disorders (10%)',
    'Moderate prevalence disorders (5%)',
    'Rare disorders (1%)'
]

# ==============================
# 4. Compute Data
# ==============================
# Beta-binomial optimized curve
beta_binomial_curve = [optimized_beta_binomial(int(n)) for n in n_beta]
asymptotic_curve = 1 / np.sqrt(np.pi * n_beta)
deduced_curve = 1/(n_beta + 1)

# Dirichlet-multinomial curves
dirichlet_exact = {k: [exact_dirichlet_multinomial(n, k) for n in n_dirichlet] 
                   for k in categories}
dirichlet_stirling = {k: [stirling_dirichlet_multinomial(n, k) for n in n_dirichlet] 
                      for k in categories}

# ==============================
# 5. Create Figure
# ==============================
plt.figure(figsize=(14, 10))

# ------------------------------
# Plot 1: Beta-Binomial Comparison
# ------------------------------
ax1 = plt.subplot(2, 1, 1)

# Plot empirical and asymptotic curves
ax1.plot(n_beta, beta_binomial_curve, 'o-', color='darkred', linewidth=2, 
         label='Optimized Beta-Binomial (max at k=n/2)')
ax1.plot(n_beta, asymptotic_curve, 'k--', linewidth=2, 
         label=r'Asymptotic: $\frac{1}{\sqrt{\pi n}}$')
ax1.plot(n_beta, deduced_curve, '^-', linewidth=2, 
         label=r'Deduced Curve')

# Add prevalence thresholds
for thresh, color, label in zip(thresholds, threshold_colors, threshold_labels):
    ax1.axhline(y=thresh, color=color, linestyle=':', alpha=0.7)
    ax1.text(51, thresh, f' {label}', va='center', fontsize=9, color=color)

ax1.set_title('Maximum Achievable Prevalence for Balanced Phenotypes (k=n/2)', fontsize=14)
ax1.set_xlabel('Number of Genes (n)', fontsize=12)
ax1.set_ylabel('Prevalence', fontsize=12)
ax1.legend(loc='upper left')
ax1.grid(alpha=0.2)
ax1.set_ylim(0, 0.6)

# Biological interpretation box
ax1.text(0.95, 0.95, 
         "Biological Insight:\nOptimizing priors maintains balanced phenotypes\n"
         "at clinically relevant prevalences (5-20%) even\n"
         "for complex polygenic systems (n=50 genes)",
         transform=ax1.transAxes, ha='right', va='top',
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray'))

# ------------------------------
# Plot 2: Dirichlet-Multinomial Comparison
# ------------------------------
ax2 = plt.subplot(2, 1, 2)

# Plot exact and Stirling approximations for each k
colors = ['blue', 'green', 'red', 'purple']
for i, k in enumerate(categories):
    # Exact values (only plot where defined)
    valid_n = [n for n in n_dirichlet if n % k == 0]
    valid_vals = [dirichlet_exact[k][n-5] for n in valid_n]
    ax2.plot(valid_n, valid_vals, 'o', color=colors[i], markersize=4,
             label=f'Exact (k={k})' if i==0 else None)
    
    # Stirling approximation
    ax2.plot(n_dirichlet, dirichlet_stirling[k], '-', color=colors[i], linewidth=2,
             label=f'Pseudo-Stirling Approx. (k={k})')

# Formatting
ax2.set_yscale('log')
ax2.set_title('Prevalence Under Neutral Prior (Dirichlet-Multinomial)', fontsize=14)
ax2.set_xlabel('Number of Genes (n)', fontsize=12)
ax2.set_ylabel('Prevalence (Log Scale)', fontsize=12)
ax2.legend(loc='lower left')
ax2.grid(True, which="both", ls="-", alpha=0.2)
ax2.set_ylim(1e-10, 1)

# Add biological interpretation
ax2.text(0.95, 0.20, 
         "Biological Insight:\nEach additional category (k) exponentially reduces\n"
         "prevalence of balanced states under neutral assumptions:\n"
         r"$\mathcal{O}(n^{(1-k)/2})$ decay → Ultra-rare disorders",
         transform=ax2.transAxes, ha='right', va='top',
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray'))

# ------------------------------
# Final Formatting
# ------------------------------
plt.tight_layout()
plt.savefig('./images/balanced_phenotype_prevalence_comparison.png', dpi=300)
plt.show()

# ==============================
# 6. Key Comparison Table (k=2)
# ==============================
print("\nPrevalence Comparison for Binary Systems (k=2):")
print("n\tOptimized Prior\tFixed Prior\tRatio")
print("-"*45)
for n in [10, 20, 30, 40, 50]:
    opt = optimized_beta_binomial(n)
    fixed = 1/(n+1)  # Dirichlet-multinomial with k=2
    print(f"{n}\t{opt:.4f}\t\t{fixed:.4f}\t\t{opt/fixed:.1f}x")
```

## The Pseudo-Stirling approximation vs. calculated Dirichlet-multinomial (alpha-multinomial) balanced probability curves.

Here, we describe the formula that appears uses an approximation (Pseudo Stirling) that reduces to a close approximation to the calculated beta-binomial curve. In this derivation a formula that would approximate the directly calculated balanced probability curves using the Dirichlet multinomial distribution for different levels of category range is considered.
The exact probability of any specific count vector (including the balanced vector) in the symmetric Dirichlet-multinomial is:

 $P_{exact}(n,k)$ = $\frac{[𝛤(n + 1)⋅𝛤(k)]}{𝛤(n + k)}\ 
 = \frac{[n!⋅(k-1)!]}{(n+k-1)!}\$
 
The Pseudo Stirling approximation:
 
is derived by applying Stirling's approximation [ 𝛤(z) ~ $(√2𝜋/z)(z/e)^{z})$] to the exact formula and simplifying for large (n) and fixed (k).

1.	Approximate Logarithmic transformation

ln$P_{app}$ ≈ (1-k)ln n + (k-1)ln(k-1) + 1/2 ln(2𝜋 (k-1))
 
2.	For large n/k, setting  
 
(1-k)/2 ln n ≈ (1-k)ln n  
1/2 ln k ≈ (k-1) ln (k-1)  
(1-k)/2 ln(2𝜋) ≈ 1/2ln(2𝜋 (k-1))  
  
The following approximation of the logarithmic Stirling approximation was contrived:

1/2 ln k + 1-k/2 ln(2𝜋)+ 1-k/2 ln n
 
Thus obtaining
 
lnP = 1/2 ln k + 1-k/2 ln(2𝜋n)

3.	Exponentiation (converting back to probability):

  P = exp (1/2 ln k + 1-k/2 ln(2𝜋n)) = $k^{1/2} ⋅2𝜋n ^{1-k/2}$
 
The relationship to beta-binomial super probability minima plot:
For k = 2, the above formula reduces to, 

 $P_{approx}$ ≈ $2^{1/2}⋅2𝜋n^{-1/2}$ = $(1/𝜋n)^{1/2}$  
  
Thus, this adjustment to the Pseudo Stirling approximation of the Dirichlet Multinomial distribution for balanced probabilities was adopted because the equation transforms seamlessly into the approximation that matches the exact calculation of the beta-binomial balanced probability curve. It should be noted that the purpose of of this process was to derive an approximate method that permitted curves that balanced between the multinomial and the uniformly distributed Dirichlet multinomial distributions (this is explained further below). It was inferred that these curves would provide the means to estimate the type and number of gene expression levels for each prevalence group. While this is entirely deductive, it is intended to assist in choosing parameters for the prevelence simulation using the Dirichlet-Multinomial distribution.
