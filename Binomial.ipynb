He proposed that the recessive allele would not be expressed as a phenotype unless it is homozygous. The generalised binomial equation that calculates the column marginal (totals at the bottom of the column) is, $^{n}C_{r~}P(r)^{r}(1-P(r))^{n-r}$, where, $^{n}C_{r}$ is the binomial coefficient that can be found on the binomial distribution table and P(r) is the top row marginal giving the dominant gene expression probability. Notice that the probabilities in the top row are all, the estimates of probability for the corresponding number of dominant alleles to be expressed in a n-gene combination. Thus, in a polygenic Mendelian combination, it is the total probability of expression of all dominant alleles that add up to 0.75. Same argument can be applied to the expression of a recessive trait (second row).
The formula for calculating the total mariginal probability of dominant expression is $ùõ¥_{r=0}^{n}(r)P(r)^{r}(1-P(r))^{n-r}$.

## Using a Heatmap to display Binomial Probability Distribution

Below, we demonstrate two different binomial probability distribution patterns, depending on if it is a symmetric distribution of probability (e.g. coin toss with all probabilities = 0.5) or a Mendelian one (0.75:0.25 distribution). First, we assume that the alleles have an equal probability of expression. The following code will plot the heat map for the probabilities. Examining the heat map will demonstrate that the maximum probability of expresion follows a relatively narrow band down the probability ladder. 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def plot_binomial_heatmap(max_genes=12):
    """Generates a heatmap of binomial probabilities for gene expression."""
    n_values = np.arange(0, max_genes + 1)  # Number of genes (0 to 12)
    k_values = np.arange(0, max_genes + 1)  # Number of dominant traits expressed (0 to 12)
    prob_matrix = np.zeros((len(n_values), len(k_values)))
    
    # Calculate probabilities
    for i, n in enumerate(n_values):
        for j, k in enumerate(k_values):
            if k <= n:  # Only possible when k <= n
                prob_matrix[i, j] = stats.binom.pmf(k, n, 0.5)  # Assuming p=0.5 for dominant traits
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(prob_matrix, annot=True, fmt=".3f", cmap="YlGnBu", 
                xticklabels=k_values, yticklabels=n_values,
                cbar_kws={'label': 'Probability'})
    plt.xlabel("Number of Dominant Traits Expressed (k)")
    plt.ylabel("Number of Genes (n)")
    plt.title("Binomial Distribution Probabilities for Gene Expression")
    plt.tight_layout()
    plt.show()

plot_binomial_heatmap()
```

If we now change the code to have the dominant probability to be 0.75, the maximum probability of expression follows a narrower band and shifted to the right (towards more dominant expression). This is of interest because Dingle etal (2015) and Dennis Noble's group have pointed out the impossibility of the human genome (with 30 000 genes) evolving without the pathways of evolution only exploring a tiny proportion of possibilities, due to those selected early, determining the evolutionary path. The above binomial pattern may provide us with a possible way to model why multiple genes may independently contribute to a narrow path of high probability. In reality, genes do not act independently, and gene interactions during evolution may narrow the prbable paths further. This idea will be explored as we gradually generalise the idea of polygenic inheritance from Mendelian to other forms including eventually includinmg gene-gene interactions.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def plot_mendelian_heatmap(max_genes=12):
    """Generates a heatmap of Mendelian probabilities for gene expression."""
    n_values = np.arange(0, max_genes + 1)  # Number of genes (0 to 12)
    k_values = np.arange(0, max_genes + 1)  # Number of dominant traits expressed (0 to 12)
    prob_matrix = np.zeros((len(n_values), len(k_values)))
    
    # Calculate Mendelian probabilities with p=0.75
    for i, n in enumerate(n_values):
        for j, k in enumerate(k_values):
            if k <= n:  # Only possible when k <= n
                prob_matrix[i, j] = stats.binom.pmf(k, n, 0.75)  # p=0.75 for dominant traits
    
    plt.figure(figsize=(12, 10))
    heatmap = sns.heatmap(prob_matrix, annot=True, fmt=".3f", cmap="YlGnBu", 
                         xticklabels=k_values, yticklabels=n_values,
                         cbar_kws={'label': 'Probability'})
    
    plt.xlabel("Number of Dominant Traits Expressed (k)")
    plt.ylabel("Number of Genes (n)")
    plt.title("Mendelian Inheritance Probabilities for Gene Expression (p=0.75)")
    plt.tight_layout()
    plt.show()
plot_mendelian_heatmap()
```

## Case of 4 genes and balance between probability of Dominant vs Recessive

We return to the previous table for probabilities of expression for the 4 gene Mendelian inheritance. The third column from the left demonstrates equal numbers of dominant and recessive alleles. This is of interest later regarding a concept called balanced-expression states. For this reason we will discuss this further for now.

<img src="./images/Mendel5.png" width=800, height=300>

The middle column above has equal porportions of dominant and recessive traits being expressed (2 D and 2 R for each 4 gene combination). There are 27 gene combinations for each of these trait combinations of all Dominant and all Recessive equal in frequency. The total probability for this column is 54/256 = 0.210 ~ 4C2(0.75)2(0.25)2 = 4!/4-2)!2! x (0.5625)(0.0625) = 6 x 0.0351.

Here, the nCr is such that r is n/2 and, the probabilities for dominant and recessive expression are equal or balanced. There are 54 pairwise combinations of alleles, with only 2 dominant traits. It is best to consider such combinations (e.g. for genes D,E,F,G, the gene combinations may be DDEEffgg, or DdEeffgg to give the phenotype DEfg), in terms of how the gene combinations that lead to the dominant phenotype would configure (the other two phenotypes would always he two homozygous combination).

Thus, we reduce the calculation to that of the number of dominant gene combinations that are possible for two genes. This number is 9. 

Next, considering 4 genes (four phenotypes), there would be 6 different ways in which these phenotypes could combine in pairs. Thus, we have the number of combinations of 2 dominant and two recessive phenotypes equal to 54.

Here, there are 27 pairwise combinations that are dominant and 27 recessives. The idea of balanced probability of dominant and recessive expression can be applied to any ratio of marginal probabilities as shall be illustrated later (e.g. 0.8:0.2 instead of 0.75:0.25). We will use this as a deductive construct to make inferences about prevalence.

The binomial distribution above does not take into account the variability of expression of say the set of ‚Äúdominant‚Äù genes (alleles), which might cause the 0.75 probability to change to a different marginal probability as suggested above.  We can attempt to vary the probabilities within the binomial distribution. However, attempting to vary the total probabilities (for dominant and recessive) from 0.75:0.25 ratio faces the overdispersion problem.

## Moving from Mendelian probabilities to variable probability of expresion

The code below attempts to model variation of the dominant probability P(r) from 0.5 to 0.8. As is seen the curves plots are quite close to each other. It does not model natural data staisfactorily because the data is overdispersed compared to the model

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, binom
from scipy.special import comb, betaln

# Set parameters
n_genes = 10

# Number of genes (trials)
alpha_priors = [n_genes*0.8, n_genes*0.75, n_genes*0.6, n_genes*0.5]  # Prior hyperparameters (Œ±)
beta_priors = [n_genes*0.2, n_genes*0.25, n_genes*0.4, n_genes*0.5]   # Prior hyperparameters (Œ≤)

# Generate data
k_values = np.arange(0, n_genes + 1)

# Plot
plt.figure(figsize=(10, 6))
colors = ['#1f77b4','#6e33ff','#ff7f0e', '#2ca02c']

for i, (a, b) in enumerate(zip(alpha_priors, beta_priors)):
    # Compute Binomial PMF
    
    pmf =[ binom.pmf(k, n_genes, a/(a+b) ) for k in k_values]

    # Plot
    plt.plot(k_values, pmf, 'o-', color=colors[i], 
             label=f'P(r) ={a/n_genes}, P(n-r) ={b/n_genes}')
    
    # Overlay pure Binomial (for comparison)

plt.title(f'Binomial- Variable Penetrance (n = {n_genes} Genes)')
plt.xlabel('Number of Dominant Traits (k)')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```
