# Preliminary Exploration of the Deductive DM model- Generating Hypotheses

## Preamble

In prior chapters, we have so far built a model of Genomics based on Mendelian inheritance followed by deviation from this using probability and statistical models. The model used first was the beta-binomial model that explored the outcomes based on variance of the modal probability of expression. This modal probability can be considered the margianl (total) probability of expression for each allele type in a bi-allelic expression pattern (one allele always being expressed with a higher probability).

Using  the beta-binomial model and using the number of dominant genes expressed as the generator of ùõº (and ùõΩ), it was possible to construct a new model called the super-probability curves. These represented the maximum probability of expression gene configurations. The minima of these super-probability curves was a convenient point to plot to estimate the minimum probability of expression of most likley range of configurations. This minimum point was always at k = N/2. Plotting of these points demonstrated that the probability of expression of even the minimum super probability reached an asymptote above 0.05 for N = 50-100. Thus it was unlikely that the beta-binomial distribution alone would satisfactorily model prevalence levels below this level. (While there were gene-configurations that had lower probability, the demonstration of most liklely evolutionary paths to phenotypes being higher probability configurations prompted maintaining this approach of pursuing the minimum limit for most probable configurations.

The next stage in exploring independent gene expression was to consider the Dirichlet-Multinomial (DM) distribution. The ùõº (concentration) parameter in this distribution determines how data concentrates. The other aspect of the DM distribution is that it involves more than two categories of expression. We found that plausible expression levels were limited to up to 5 (again consitent with empirical studies), because this curve reached the lower threshold of prevalence of 0.1% below which, genetic expression is autosomal heterozygous abnormalities (introns and indels).

## Generating Hypotheses

Following some heuristic trials, it was evident that the pattern of curves that was most suitable for considering the interaction between expression levels, gene numbers and prevalence was a balanced expression with ùõº at 1 (probability difference between this and ùõº < 1 and ùõº > 1 were minimal to affect prevalence estimates and it made calculations easier) and to use Stirling approximation as indicated in the previous section. A further, attraction of this approach was that the Stirling approximation for the DM model matched the approximation that closely matched the exactly calculated beta binomial super probability curve minima. Thus, it was accepted on heuristic grounds, that this exercise in considering the DM model of gene expression is a valid model to model prevalence levels of diseases and thus had sufficient deductive validity to continue with further simulation of prevalence levels.

It should be noted that the fact that Stirling approximation curves fit a less dispersed range and at higher probability levels point to ùõº > 1 being biologically more plausible. The code below demonstrate how the probability plots as we vary ùõº from 0.1 to 2 raises the curve which is correspondingly less dispersed.

```python
import numpy as np
from scipy.special import gammaln, loggamma
import matplotlib.pyplot as plt

def dm_balanced_log_prob(n, k, alpha):
    """Log probability of perfectly balanced outcome in Dirichlet-Multinomial"""
    if n % k != 0:
        return -np.inf  # Only defined when n divisible by k
    
    d = n // k  # Count per category
    alpha0 = k * alpha
    
    # Log probability calculation
    log_prob = (gammaln(n + 1) + gammaln(alpha0) - gammaln(n + alpha0))
    log_prob += k * (gammaln(d + alpha) - gammaln(d + 1) - gammaln(alpha))
    return log_prob

def stirling_approx_log_prob(n, k, alpha):
    """Stirling approximation of log probability for balanced outcome"""
    d = n // k
    log_prob = -0.5 * (k - 1) * np.log(2 * np.pi * n)
    log_prob -= 0.5 * k * np.log(2 * np.pi * d)
    log_prob += 0.5 * np.log(k)
    log_prob += (alpha0 - 0.5) * np.log(alpha0)
    log_prob -= (n + alpha0 - 0.5) * np.log(n + alpha0)
    log_prob += k * ((d + alpha - 0.5) * np.log(d + alpha) - (alpha - 0.5) * np.log(alpha))
    return log_prob

# Parameters
n = 100  # Total observations
k_values = np.arange(2, 101)  # Number of categories from 2 to 100
alpha_values = [0.1, 0.5, 1.0, 1.5, 2.0]  # Different concentration parameters

# Compute and plot
plt.figure(figsize=(12, 8))

for alpha in alpha_values:
    log_probs = []
    valid_k = []
    
    for k in k_values:
        if n % k == 0:
            log_probs.append(dm_balanced_log_prob(n, k, alpha))
            valid_k.append(k)
    
    # Normalize and convert to probability scale
    max_log_prob = max(log_probs)
    probs = np.exp(log_probs - max_log_prob)
    
    plt.plot(valid_k, probs, label=f'Œ± = {alpha}', linewidth=2.5)

plt.xlabel('Number of Categories (k)', fontsize=14)
plt.ylabel('Relative Probability of Balanced Outcome', fontsize=14)
plt.title('Dirichlet-Multinomial Balanced Outcome Probability\n(n=100 total observations)', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.yscale('log')
plt.xlim(2, 100)
plt.tight_layout()
plt.show()
```

The family of curves in the figure below are the Striling approximation of the DM model that illustrates that DM distribution results in delineation of different prevalence classes for polygenic cases, in a range from a little over 0.2 probability to 0.01 probability. The curves further delineate these prevalence levels, according to ranges found in empirical disorders.

```python
import numpy as np
from scipy.special import gammaln, gamma
import matplotlib.pyplot as plt

def continuous_prevalence(n, categories=3):
    """
    Models prevalence as a continuous function using:
    P ‚âà (categories)^(1/2) * (2œÄn)^((1-categories)/2)
    Derived from Stirling's approximation of Dirichlet-Multinomial
    """
    # Stirling's approximation for continuous n
    if n <= 0:
        return 0
    
    # Main term: (categories)^(1/2) * (2œÄn)^((1-categories)/2)
    log_prevalence = 0.5 * np.log(categories) + (1 - categories)/2 * np.log(2 * np.pi * n)
    return np.exp(log_prevalence)

# Parameters
gene_counts = np.linspace(5, 200, 500)  # More points for smoother curve
categories = [2, 3, 4, 5]

# Compute prevalences
results = {c: [continuous_prevalence(n, c) for n in gene_counts] for c in categories}

# Plotting
plt.figure(figsize=(14, 10))
colors = plt.cm.viridis(np.linspace(0, 1, len(categories)))

# Plot smooth curves
for i, c in enumerate(categories):
    plt.semilogy(gene_counts, results[c], '-', color=colors[i], 
                 label=f'{c} expression levels', linewidth=3)

# Threshold definitions
thresholds = [
    (0.05, 'green', 'Common (>5%)'),
    (0.01, 'orange', 'Moderate (1-5%)'),
    (0.001, 'red', 'Rare (0.1-1%)'),
    (0.0001, 'purple', 'Ultra-rare (<0.01%)')
]

# Add threshold lines
for thresh, color, label in thresholds:
    plt.axhline(y=thresh, color=color, linestyle='--', alpha=0.7, linewidth=1.5)
    plt.text(10, thresh * 1.2, label, color=color, fontsize=11,
             bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

# Add critical points with annotations
for i, c in enumerate(categories):
    for thresh, color, label in thresholds:
        probs = np.array(results[c])
        below_thresh = np.where(probs < thresh)[0]
        
        if len(below_thresh) > 0:
            n_crit = gene_counts[below_thresh[0]]
            p_val = probs[below_thresh[0]]
            
            # Plot marker
            plt.plot(n_crit, p_val, 'o', markersize=10, color=color,
                     markeredgecolor='white', markeredgewidth=1.5, zorder=10)
            
           
            
            # Add annotation
            plt.annotate(f'{n_crit:.1f} genes', (n_crit, p_val),
                         xytext=(n_crit + 8, p_val * (3 if thresh > 0.001 else 0.3)),
                         fontsize=10, color=color,
                         arrowprops=dict(arrowstyle="->", color=color, linewidth=1.5))

# Formatting
plt.xlabel('Number of Genes in Pathway', fontsize=14, fontweight='bold')
plt.ylabel('Prevalence of Balanced Expression', fontsize=14, fontweight='bold')
plt.title('Prevalence of Balanced Expression Phenotypes\n(Continuous Approximation Model)', 
          fontsize=16, fontweight='bold', pad=20)

plt.legend(fontsize=12, loc='upper right')
plt.grid(True, which="both", ls="-", alpha=0.2)
plt.ylim(1e-7, 1)
plt.xlim(5, 200)

plt.tight_layout()
plt.savefig('./images/continuous_prevalence_model.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Conclusions from the DM Stirling Curves

Adopting the Dirichlet9Alpha)-Multinomial model permits exploration of multiallelic type gene expression and interactions. It allows the modelling of a range of prvalence levels that are closer to what is found in postulated polygenic disorders. While acknowledging that the polygenic expression is yet independent in this statistical model, it is useful to explore its fit with expected prevalence levels. The curves above provide some conclusions:

1)	The **high prevalence disorders** (5-20%) are hypothesised to involve 2 expression levels in polygenic configurations. At most, there may be 3-4 levels/variants of expression in a minority of genes. What cannot be known at this stage is if such genes would be core genes or only peripheral or regulatory genes. This hypothesis needs to be tested. However, in a model of dependent relations, 3-4 levels of expression in a small number of genes may be due to regulatory alteration of gene expresion. Possible mechanisms are discussed in the next section.
  
2)	The **medium level prevalence disorders** at 1-5% (bipolar mood disorder, ADHD) are most likely induced by genes that are expressed at 3 levels/variants of expression with 4-5 levels of expression in a small number of genes.
  
3)	The **rare disorders** at 0.1-1% (schizophrenia) are likely to be induce by complex gene expression patterns at 4-5 levels/variants of expression (significant proportions of these curves are found within the range of rare disorders.

It is the preliminary conclusion of this study that high prevalence disorders such as depression and chronic pain are more susceptible and prevalent because only 2 expression levels (variants) need to be induced to manifest the conditions. It is conjectured that the mechanisms would be mostly due to epigenetic effects. It would be interesting to consider if rare condition such as schizophrenia, with its known resistance to pain may have protections afforded by genes expressed at multiple levels resulting in complex effects on brain networks such that any sensitivity to pain is submerged. Identifying such gene environment interactions may be a line of investigation to be considered in genomics.

Finally, with regard to diseases with prevalence levels below 0.1%, they are considered to be autosomal (heterozygous) in nature and have high likelihood of expression and penetrance. They are not likely to be modelled by the polygenic model. However, it would be possible to include such disorders within the model by excluding them from the DM calculations but adding them to the overall disease profile after DM-calculations. Below, these parameters developed during empirical genomic studies are further elaborated. 


### Synergy

Synergy is the impact that genes have on each other producing probability of expression jointly that is greater than the sum of probabilities. With higher synergy there is more likelihood of genes activating each other. Synergy Ratio is calculated using the observed OR divided by the products of the individual Odds Ratio (OR) [the predicted OR]. Thus the Synergy Ratio is greater than 1. In order to represent real world SF in this model, it would be necessary to produce the following relation SF (model) = log SF (real). This transformation would be important as the parameters need to range between 0 and 1. An interesting outcome is that products will be transformed into sums. The outcome would be similar. 

### Environment Noise

Environmental effects would be mediated by epigenetic effects that may or may not have transgenerational expression. These effects may be expressed in terms of inhibition (methylation) or activation (acetylation) of genes. Environmental effects as a parameter would be expected to either oppose or aid the impact of synergy. The genotype x environment interactions GxE is an important factor to consider between synergy and environment. Environmental noise is measured by dividing the standard deviation by mean (standardization or normalization).

### Threshold Ratio

Threshold ratio for polygenic instances is calculated using Polygenic Risk Scores (PRS) calculated as a threshold. For common, moderate, rare disorders these thresholds should be respectively, in the order of 0.1, 0.05. 0.01. Linkage Disequilibrium (LD) is an interesting method for measuring or fine-tuning threshold ratio as well. It is possible to use measures of LD could be useful to incorporating in a hierarchical graph (see later), when taking account of SNP data.

### Penetrance

Penetrance is the probability of expression of the phenotype in an individual genome. Penetrance is closely related to population prevalence. 
Population prevalence = frequency of the genotype in the population x penetrance.
In the language of probability penetrance is P(D|G)  the probability that an individual with a specific genotype will produce the phenotype D where P(D) or prevalence is the proportion of a population that expresses a particular phenotype D and P(G) is the allele frequency of a particular genotype in the population.

**P(D) = P(D|G) * P(G)**
